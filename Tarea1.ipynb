3. Hacer una tabla donde la primera columna sea el mes, la segunda el índice
Id (por trampa) y la tercera el índice Cx para el año 2015.

```python
#Numero de trampas instaladas
df_2015.drop_duplicates(subset=['No_Trampas_Instaladas'])
```

```python
#Indice de disperción: id=var/media muestral
#Coeficiente de Green: Cx=[(var/media muestral)-1]/[Suma x-1] 

##################################################Enero
Enero = df_2015[df_2015['Mes']==1] 
ENT1 = Enero[Enero['No_Trampas_Instaladas']==0]
ENT2 = Enero[Enero['No_Trampas_Instaladas']==1]
TotalEnero = pd.concat([ENT1, ENT2]) 
#Id
A1=TotalEnero['No_DiaphorinasXtrampa'].var()
B1=TotalEnero['No_DiaphorinasXtrampa'].mean()
IDE=A1/B1
#Cx
C1=TotalEnero['No_DiaphorinasXtrampa'].sum()
CXE=(IDE-1)/(C1-1)

############################################Febrero
Febrero = df_2015[df_2015['Mes']==2] 
FNT1 = Febrero[Febrero['No_Trampas_Instaladas']==0]
FNT2 = Febrero[Febrero['No_Trampas_Instaladas']==1]
TotalFebrero = pd.concat([FNT1, FNT2]) 
#Id
A2=TotalFebrero['No_DiaphorinasXtrampa'].var()
B2=TotalFebrero['No_DiaphorinasXtrampa'].mean()
IDF=A2/B2
#Cx
C2=TotalFebrero['No_DiaphorinasXtrampa'].sum()
CXF=(IDF-1)/(C2-1)

############################################Marzo
Marzo = df_2015[df_2015['Mes']==3] 
MARNT1 = Marzo[Marzo['No_Trampas_Instaladas']==0]
MARNT2 = Marzo[Marzo['No_Trampas_Instaladas']==1]
TotalMarzo = pd.concat([MARNT1, MARNT2]) 
#Id
A3=TotalMarzo['No_DiaphorinasXtrampa'].var()
B3=TotalMarzo['No_DiaphorinasXtrampa'].mean()
IDMAR=A3/B3
#Cx
C3=TotalMarzo['No_DiaphorinasXtrampa'].sum()
CXMAR=(IDMAR-1)/(C3-1)

##########################################Abril
Abril = df_2015[df_2015['Mes']==4] 
ANT1 = Abril[Abril['No_Trampas_Instaladas']==0]
ANT2 = Abril[Abril['No_Trampas_Instaladas']==1]
TotalAbril = pd.concat([ANT1, ANT2]) 
#Id
A4=TotalAbril['No_DiaphorinasXtrampa'].var()
B4=TotalAbril['No_DiaphorinasXtrampa'].mean()
IDABR=A4/B4
#Cx
C4=TotalAbril['No_DiaphorinasXtrampa'].sum()
CXABR=(IDABR-1)/(C4-1)

#########################################Mayo
Mayo = df_2015[df_2015['Mes']==5] 
MAYNT1 = Mayo[Mayo['No_Trampas_Instaladas']==0]
MAYNT2 = Mayo[Mayo['No_Trampas_Instaladas']==1]
TotalMayo = pd.concat([MAYNT1, MAYNT2]) 
#Id
A5=TotalMayo['No_DiaphorinasXtrampa'].var()
B5=TotalMayo['No_DiaphorinasXtrampa'].mean()
IDMAY=A5/B5
#Cx
C5=TotalMayo['No_DiaphorinasXtrampa'].sum()
CXMAY=(IDMAY-1)/(C5-1)

########################################Junio
Junio = df_2015[df_2015['Mes']==6] 
JUNNT1 = Junio[Junio['No_Trampas_Instaladas']==0]
JUNNT2 = Junio[Junio['No_Trampas_Instaladas']==1]
TotalJunio = pd.concat([JUNNT1, JUNNT2]) 
#Id
A6=TotalJunio['No_DiaphorinasXtrampa'].var()
B6=TotalJunio['No_DiaphorinasXtrampa'].mean()
IDJUN=A6/B6
#Cx
C6=TotalJunio['No_DiaphorinasXtrampa'].sum()
CXJUN=(IDJUN-1)/(C6-1)

############################################Julio
Julio = df_2015[df_2015['Mes']==7] 
JULNT1 = Julio[Julio['No_Trampas_Instaladas']==0]
JULNT2 = Julio[Julio['No_Trampas_Instaladas']==1]
TotalJulio = pd.concat([JULNT1, JULNT2]) 
#Id
A7=TotalJulio['No_DiaphorinasXtrampa'].var()
B7=TotalJulio['No_DiaphorinasXtrampa'].mean()
IDJUL=A7/B7
#Cx
C7=TotalJulio['No_DiaphorinasXtrampa'].sum()
CXJUL=(IDJUL-1)/(C7-1)

############################################Agosto
Agosto = df_2015[df_2015['Mes']==8] 
AGNT1 = Agosto[Agosto['No_Trampas_Instaladas']==0]
AGNT2 = Agosto[Agosto['No_Trampas_Instaladas']==1]
TotalAgosto = pd.concat([AGNT1, AGNT2]) 
#Id
A8=TotalAgosto['No_DiaphorinasXtrampa'].var()
B8=TotalAgosto['No_DiaphorinasXtrampa'].mean()
IDAG=A8/B8
#Cx
C8=TotalAgosto['No_DiaphorinasXtrampa'].sum()
CXAG=(IDAG-1)/(C8-1)

#############################################Septiembre
Septiembre = df_2015[df_2015['Mes']==9] 
SEPNT1 = Septiembre[Septiembre['No_Trampas_Instaladas']==0]
SEPNT2 = Septiembre[Septiembre['No_Trampas_Instaladas']==1]
TotalSeptiembre = pd.concat([SEPNT1, SEPNT2]) 
#Id
A9=TotalSeptiembre['No_DiaphorinasXtrampa'].var()
B9=TotalSeptiembre['No_DiaphorinasXtrampa'].mean()
IDSEP=A9/B9
#Cx
C9=TotalSeptiembre['No_DiaphorinasXtrampa'].sum()
CXSEP=(IDSEP-1)/(C9-1)

###############################################Octubre
Octubre = df_2015[df_2015['Mes']==10] 
OCNT1 = Octubre[Octubre['No_Trampas_Instaladas']==0]
OCNT2 = Octubre[Octubre['No_Trampas_Instaladas']==1]
TotalOctubre = pd.concat([OCNT1, OCNT2]) 
#Id
A10=TotalOctubre['No_DiaphorinasXtrampa'].var()
B10=TotalOctubre['No_DiaphorinasXtrampa'].mean()
IDOC=A10/B10
#Cx
C10=TotalOctubre['No_DiaphorinasXtrampa'].sum()
CXOC=(IDOC-1)/(C10-1)

########################################Noviembre
Noviembre = df_2015[df_2015['Mes']==11] 
NOVNT1 = Noviembre[Noviembre['No_Trampas_Instaladas']==0]
NOVNT2 = Noviembre[Noviembre['No_Trampas_Instaladas']==1]
TotalNoviembre = pd.concat([NOVNT1, NOVNT2]) 
#Id
A11=Noviembre['No_DiaphorinasXtrampa'].var()
B11=Noviembre['No_DiaphorinasXtrampa'].mean()
IDNOV=A11/B11
#Cx
C11=TotalNoviembre['No_DiaphorinasXtrampa'].sum()
CXNOV=(IDNOV-1)/(C11-1)

#############################################Diciembre
Diciembre = df_2015[df_2015['Mes']==12] 
DICNT1 = Diciembre[Diciembre['No_Trampas_Instaladas']==0]
DICNT2 = Diciembre[Diciembre['No_Trampas_Instaladas']==1]
TotalDiciembre = pd.concat([DICNT1, DICNT2]) 
#Id
A12=Diciembre['No_DiaphorinasXtrampa'].var()
B12=Diciembre['No_DiaphorinasXtrampa'].mean()
IDDIC=A12/B12
#Cx
C12=TotalDiciembre['No_DiaphorinasXtrampa'].sum()
CXDIC=(IDDIC-1)/(C12-1)
```

```python

from tabulate import tabulate
MESES = [ 
        ["Enero",IDE,CXE], 
        ["Febrero",IDF,CXF],
         ["Marzo",IDMAR,CXMAR], 
        ["Abril",IDABR,CXABR],
         ["Mayo",IDMAY,CXMAY], 
        ["Junio",IDJUN,CXJUN],
         ["Julio",IDJUL,CXJUL], 
        ["Agosto",IDAG,CXAG], 
        ["Septiempbre",IDSEP,CXSEP],
        ["Octubre",IDOC,CXOC], 
        ["Noviembre",IDNOV,CXNOV], 
        ["Diciembre",IDDIC,CXDIC]] 
columnas = ["MESES", "ID", "Cx"]

print(tabulate(MESES, headers=columnas))
```
